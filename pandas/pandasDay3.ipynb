```python
import pandas as pd
df = pd.read_csv("students.csv")
df.rename(columns={"Marks":"Score"}, inplace=True)
df
```

```python
df.drop("Attendance", axis=1, inplace=True)
df
```

```python
print(df["Score"].dtypes)
df["Score"] = df["Score"].astype(int)
df["Score"].dtypes
```

```python
print(df["Class"].dtypes)
df["Class"] = df["Class"].astype(str)
df["Class"].dtypes
```

```python
def fun(x):
    if x>=50:
        return "Pass"
    else:
        return "Fail"
df["Result"]= df["Score"].apply(fun)
df
```

```python
def fun(x):
    if x>=90:
        return "A"
    elif x<90 and x>=80:
        return "B"
    elif x<80 and x>=70:
        return "C"
    elif x<70 and x>=60:
        return "D"
    elif x<60 and x>=40:
        return "E"
    else:
        return "F"
df["Grade"]= df["Score"].apply(fun)
df
```

```python
x = [-1,40,60,75,100]
y = ["Fail","Average","Good","V Good"]

df["Performance"] = pd.cut(
    df["Score"], 
    bins=x, 
    labels=y
)
df
```

```python
df["Performance"].value_counts()
```

```python
df.groupby("Subject")["Score"].agg(["mean","max","min","count"])
```
