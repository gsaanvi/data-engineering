```python
#stacking arrays
import numpy as np
a = np.array([1,2])
b = np.array([3,4])
c = np.array([5,6])
d = np.vstack((a, b, c)) #vertical
e = np.hstack((a, b, c)) #horizontal
print(d,"\n")
print(e)
```

```python
#flattening 2D array
q = np.array([[1,2], [3,4], [5,6]])
p = q.flatten()
print("Before flattening: \n", q)
print("\n After flattening:",p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = np.arange(1,8)
temperature = np.array([30,22,27,32,30,40,29])
plt.plot(days, temperature, marker='o')
plt.title("Weekly Temperature Report")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.show()
```

```python
students = np.arange(1,11)
marks = np.array([87,60,90,72,88,92,81,65,95,99])

plt.bar(students, marks)
plt.title("Student vs Marks")
plt.xlabel("Student Roll Number")
plt.ylabel("Marks")
plt.show()
```
