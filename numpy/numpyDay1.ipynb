```python
#Creating a NumPy array with 6 numbers
import numpy as np
A = np.array([1,2,3,4,5,6])
print(A)
```

```python
#array properties -> dimension, structure, size, data type
print(A.ndim)  
print(A.shape)
print(A.size)
print(A.dtype)
```

```python
#creating matrix
m = np.array([[20,40],
              [50,70]])
print(m[0,:])
print(m[:,0])
```

```python
print(A[-1])
print(A[0:5:2])
```

```python
print(A * 3) #multiplying number to numpy array
```

```python
#1D->2D
import numpy as np
arr = np.array([1,6,8,4,2,5,7])
print("1D ARRAY: ")
print(arr,"\n")

print("1D->2D ARRAY: ")
newArr = arr[:, np.newaxis]
print(newArr)
```
