```python
import numpy as np
arr = np.array([5,12,9,23,4]) 
print(arr[arr<10])
```

```python
arr = np.random.randint(0,101,size=10)
print(arr)
a = np.where(arr >=50,1,0)
print(a)
```

```python
import numpy as np
import matplotlib.pyplot as plt
arr = np.random.rand(20)
print(arr)

x = np.linspace(0,21,20)
y = np.random.rand(20)
plt.plot(x, y)
plt.title("random 20")
plt.xlabel("index")
plt.ylabel("random 0-1")
plt.show()
```

```python
a = np.array([[1,2,3], [4,5,6]])
b = np.array([[7,8,9], [10,11,12]])
c = np.concatenate((a, b), axis=0)
print(c)
print("\nshape:",c.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(1, 11, 1000)

plt.hist(x, bins=10)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Random Integers (110)")
plt.show()
```
