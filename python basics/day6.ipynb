```python
#Square Function
def square(x):
    return x*x
a = square(10)
print(a)
```

```python
#SimpleCalculator
def calc(a, b, op): 
    if op == "+": 
        return a + b 
    elif op == "-": 
        return a - b 
    elif op == "*": 
        return a * b 
    elif op == "/": 
        return a / b
print(calc(10, 15, "+"))
```

```python
#Even/Odd Checker
def numChecker(x):
    if x%2==0:
        print("Even")
    else:
        print("Odd")
numChecker(19)
numChecker(2)
```

```python
#Use *args
def sum(*args):
    total=0
    for x in args:
        total += x
    return total

res = sum(1, 2, 5)
print(f"The sum of 1, 2, and 5 is: {res}")
```

```python
#Use **kwargs
def details(**kwargs):
    print(kwargs)
details(name="Noor", course="B.Tech", age=20)
```

```python
#Use filter() &lambda
num =[1,2,4,5,7,8,9,10]
even= list(filter(lambda x:x%2==0, num))
print(even)
```

```python
#Use map() &lambda
num = [1, 2, 3, 4, 5]
double = list(map(lambda x: x * 2, num)) 
print(double)
```
