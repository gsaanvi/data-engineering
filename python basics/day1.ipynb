```python
#Basic Calculator
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Select Operator: '+', '-', '*', '/'")

if op == '+':
    print("a+b =", a + b)
elif op =='-':
    print("a-b =", a - b)
elif op == '*':
    print("a*b =", a*b)
elif op == '/':
    print("a/b =", a/b)
else:
    print("Invalid Operation! Try again..")
    
```
