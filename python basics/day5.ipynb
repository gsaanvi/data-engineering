```python
#Positive/Negative/Zero
a = int(input("Enter a number: "))
if a==0:
    print(a)
elif a>0:
    print(a, "is positive")
else:
    print(a, "is negative")
```

```python
#Grade System
marks= int(input("enter marks: "))
if marks>=75:
    print('A')
elif marks>=60 and marks<75:
    print('B')
elif marks>=45 and marks<60:
    print('C')
else:
    print("Fail")
```

```python
#Reverse Loop
for i in range(10,0, -1):
    print(i)
```

```python
#Even Number Printer
for i in range(1,31):
    if i%2==0:
        print(i)
```

```python
#Multiplication Table
num = int(input("Enter number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num*i}")
```
