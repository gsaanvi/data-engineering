```python
#Reverse a String
string = input("Enter some string: ")
print(string[::-1])
```

```python
#Count CharacterOccurrence
word = input("Enter word: ")
ch = input("Enter character: ")

print(f"{word} has {word.count(ch)} {ch}s")
```

```python
#Convert Sentence → List
sentence = input("Enter a sentence: ")
print(sentence.split())
```

```python
#Join Word List →Sentence
ls = ["I", "am", "a", "student"]
print(" ".join(ls))
```

```python
#Sort Numbers
num = []
for i in range(5):
    a = int(input("Enter number: "))
    num.append(a)
print("list before sorting : ", num)
num.sort()
print("sorted list: ", num)
```

```python
#Filter EvenNumbers
num=[2,57,8,12,45,20]
for i in range(len(num)):
    if num[i]%2==0: 
        print(num[i], end=" ")
```
