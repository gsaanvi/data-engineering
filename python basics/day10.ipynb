```python
#creating a class & objects
class Student:
    def __init__(self, name, roll):
        self.name=name
        self.roll=roll
    
ob1= Student("Noor", 47)
ob2= Student("Tara", 62)
print(ob1.name, ob1.roll)
print(ob2.name, ob2.roll)
```

```python
#using constructor to initialize & method to display
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print("Employee Name:", self.name)
        print("Salary:", self.salary)

em = Employee("Rahul", 50000)
em.show_details()
```

```python
#class attribute vs instance attribute
class College:
    branch="CSE"  
    def __init__(self, name, cgpa):
        self.name=name
        self.cgpa=cgpa

a=College("Rahul", 7.2)
b=College("Riya", 8.9)

print(a.name, a.cgpa, a.branch)
print(b.name, b.cgpa, b.branch)
```

```python
#parent-child class inheritance
class Person:
    def display(self):
        print("This is inheritance")
class Student(Person):
    pass
s=Student()
s.display()
```

```python
#overriding
class Food:
    def type(self, name):
        self.name=name
        print(self.name, "is edible")
        
class Veggie(Food):
    def type(self, name):
        self.name=name
        print(self.name, "is a veggie")
        
f=Veggie()
f.type("Potato")
```

```python
#private variable
class Data:
    def __init__(self, passwords):
        self.__passwords=passwords
        
    def show_pass(self):
        print(self.__passwords)

p=Data("qwnbt@34")
p.show_pass()
```

```python
#protected member
class Animal:
    def __init__(self):
        self._type = "bird"


class Sparrow(Animal):
    def tell_type(self):
        print(self._type)

s = Sparrow()
s.tell_type()
```

```python

#runtime polymorphism
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

animals = [Dog(), Cat(), Animal()]
for a in animals:
    a.speak()
```

```python
#IS-A and HAS-A relationships
class Engine:
    def start(self):
        print("Engine started")

class Vehicle:
    def move(self):
        print("Vehicle is moving")


class Car(Vehicle):       
    def __init__(self):
        self.engine = Engine() 

    def drive(self):
        self.engine.start()
        print("Car is driving")

c = Car()
c.move()
c.drive()
```

```python
#super keyword

class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  
        self.age = age


c = Child("Rahul", 20)
print(c.name)
print(c.age)
```

```python
#multiple inheritance (diamond problem)

class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")

class C(A):
    def show(self):
        print("Class C")

class D(B, C):
    pass

d = D()
d.show()
```
